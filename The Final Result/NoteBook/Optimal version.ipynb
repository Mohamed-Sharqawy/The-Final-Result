<h3>Handling (product price) (product minimum offer price) (product original price)</h3>
<p>This Process applies to Three of them to handle inapropriate datatype:<br>
1- having ($) and (,) in a value wont allow the type converting:<br>
    using the replace function to replace'em with ('') .<br>
    looping throug the data to avoid the nulls as they read by default as float values and the replace function wont work.<br>
2- converting the column data type to float to handle the nulls. </p>

```python
for index, value in df['product_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_price'] = value.replace('$','')
```

```python
for index, value in df['product_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_price'] = value.replace(',','')
```

```python
df['product_price'] = df['product_price'].astype('float')
```

```python
for index, value in df['product_original_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_original_price'] = value.replace('$','')
```

```python
for index, value in df['product_original_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_original_price'] = value.replace(',','')
```

```python
df['product_original_price'] = df['product_original_price'].astype('float')
```

```python
for index, value in df['product_minimum_offer_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_minimum_offer_price'] = value.replace(',','')
```

```python
for index, value in df['product_minimum_offer_price'].items():
    dt = type(value)
    if dt == float:
        pass
    else:
        df.at[index, 'product_minimum_offer_price'] = value.replace('$','')
```

```python
df['product_minimum_offer_price'] = df['product_minimum_offer_price'].astype('float')
```

<h3>Identifying if the data has outliers</h3>
This step is necessary for the replacing the nulls wether we gonna use the (Mean) or (Median).

```python
Q1 = df['product_price'].quantile(0.25)
Q3 = df['product_price'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = df[(df['product_price'] < Q1 - threshold * IQR) | (df['product_price'] > Q3 + threshold * IQR)]
outliers
```

```python
df.describe()
```

<h3>Replacing the Nulls</h3>
Given That The data has Outliers so I chose to replace the Nulls With Median

```python
df.fillna({'product_price':df['product_price'].median()}, inplace = True)
```

```python
df.fillna({'product_star_rating':df['product_star_rating'].median()}, inplace = True)
```

```python
df.fillna({'product_minimum_offer_price':df['product_minimum_offer_price'].median()}, inplace = True)
```
